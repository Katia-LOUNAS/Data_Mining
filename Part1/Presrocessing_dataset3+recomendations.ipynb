```python
df = data[['Soil', 'Crop', 'Fertilizer', 'Rainfall']]

# Convert each row to a set and store in a DataFrame
new_df = pd.DataFrame({'transaction': [set(row) for _, row in df.iterrows()]})

# Display the resulting DataFrame
new_df
```

```python
data = []
for index, row in new_df.iterrows():
    transaction = row['transaction']
    data.append(set(transaction))
min_support = 5
min_confidence = 0.5
min_correlation = 0.1
association_rules, nombre_itemsets, nomber_regles = appriori.regles_d_association(data, min_support, min_confidence, min_correlation, "confidence")
print(f"Nombre d'itemsets fréquents: {nombre_itemsets}")
print(f"Nombre de règles d'association: {nomber_regles}")
for rule in association_rules:
    A, B, confidence, sup= rule
    print(f"{list(A)} => {list(B)} - Conf: {confidence}")
```

```python
input_data = ['alluvial', 'DAP'] 

matching_sets_B = []
for rule in association_rules:
    A, B, confidence, sup = rule
    if set(input_data) == A:
        matching_sets_B.append(list(B))

if matching_sets_B:
    print(f"Input: {input_data} matches rules:")
    for B_set in matching_sets_B:
        print(f"Return set B: {B_set}")
else:
    print(f"No matching rule found for input: {input_data}")
```
