```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
class processing:
    
    def load_data(self, path):
        data = pd.read_csv(path)
        return(data)
    
    def print_info(self, data):
        print("The data is composed of " + data.shape[0] + "rows and " + data.shape[1] + "columns")
        print("The attribute list for this dataset is :", data.columns)
        print(data.head())
        
    def central_trend(self, data):
        # get the data shape
        n = data.shape[0]
        m = data.shape[1]
        # Iterate through each attribute to calculate the info
        for i in range(m-1):
            # Sorting the column values
            sorted_data = data.iloc[:, i].sort_values(ascending=True)
            #print(sorted_data)
            # Calculate the median
            if m % 2 == 0:
                median = (sorted_data.iloc[m // 2 - 1] + sorted_data.iloc[m // 2]) / 2
            else:
                median = sorted_data.iloc[m // 2]
            print(data.columns[i] + ':' )
            print("Median:", median)

            # Calculate the average (moyenne)
            moyenne = np.sum(sorted_data) / len(sorted_data)
            print("Mean:", moyenne)

            # Calculate mode
            frequency = {}

            # Iterate the data 
            for value in sorted_data:
                if value in frequency:
                    frequency[value] += 1
                else:
                    frequency[value] = 1

            # Find the value with the highest frequency
            modes = []
            max_frequency = 0

            for value, frequency in frequency.items():
                if frequency > max_frequency:
                    modes = [value]
                    max_frequency = frequency
                elif frequency == max_frequency:
                    modes.append(value)
            print("Modes values:", modes)
    
    def quartile(self, column_data):
        # Sorting the column values
        sorted_data = column_data.sort_values(ascending=True)

        # Dictionary to store quartile values
        quartile_dict = {}

        # Calculate Q0 (min)
        Q0 = sorted_data.iloc[0]
        quartile_dict['Q0'] = Q0

        # Calculate Q1
        n = len(sorted_data)
        if n % 4 == 0:
            Q1 = (sorted_data.iloc[n // 4 - 1] + sorted_data.iloc[n // 4]) / 2
        else:
            Q1 = sorted_data.iloc[n // 4]
        quartile_dict['Q1'] = Q1

        # Calculate Q2 (median)
        if n % 2 == 0:
            Q2 = (sorted_data.iloc[n // 2 - 1] + sorted_data.iloc[n // 2]) / 2
        else:
            Q2 = sorted_data.iloc[n // 2]
        quartile_dict['Q2'] = Q2

        # Calculate Q3
        if n % 4 == 0:
            Q3 = (sorted_data.iloc[3 * (n // 4) - 1] + sorted_data.iloc[3 * (n // 4)]) / 2
        else:
            Q3 = sorted_data.iloc[3 * (n // 4)]
        quartile_dict['Q3'] = Q3

        # Calculate Q4 (max)
        Q4 = sorted_data.iloc[n - 1]
        quartile_dict['Q4'] = Q4

        # Print the results
        print(f"Quartile values for the column:\n{quartile_dict}")

        return quartile_dict         #print(result_dict['Column']['Q1'])  to access to the quaetile
    
    def missing_value(self, column_data):
        # Dictionary to store missing values and their percentages for the column
        missing_values_dict = {}

        # get the data shape
        n = len(column_data)

        # Calculate the number of missing values
        nb_valeurs_manquantes = column_data.isnull().sum()

        # Calculate the percentage of missing values
        pourcentage_valeurs_manquantes = (nb_valeurs_manquantes / n) * 100

        # Store the information in the dictionary
        missing_values_dict['Number of missing values'] = nb_valeurs_manquantes
        missing_values_dict['Percentage of missing values'] = pourcentage_valeurs_manquantes

        # Print the information
        print(f"Column: {column_data.name}\nNumber of missing values: {nb_valeurs_manquantes}\nPercentage of missing values: {pourcentage_valeurs_manquantes:.2f}%\n")

        return missing_values_dict     #result_dict['Column']['Percentage of missing values']
    
    def Scatter_plot(self, data, att1 , att2):
    
        plt.scatter(data[att1],data[att2], marker='o')

        plt.title("Scatter plot")
        plt.xlabel(att1)
        plt.ylabel(att2)
        
    def histogramme(self, data , attribut):
        plt.hist(data[attribut], bins=10, color='b', edgecolor='k')
        plt.xlabel("X")
        plt.ylabel("Y")
        plt.title("Histogramme")
        
    def box_plot(self, attribut):
        plt.boxplot(attribut)
        plt.title("Box")
        plt.show












        
        
```
